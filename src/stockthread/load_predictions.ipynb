```python
path='/root/pythondev/JanePython/'

outputfile = 'stocks2.csv'
savepath=path+outputfile

data=pd.read_csv(savepath,parse_dates=['Date'],index_col='Date')
if(data.isna().all().sum()>0):
    data=data[data.columns[data.isna().all()]].dropna(axis=1,inplace=True)
data.columns[data.isna().any()].any()
data=data[data.columns[data.isna().sum()==0]]
data=data.resample('C').mean()
```

```python
import pandas_datareader.data as web
import datetime
daily_tickers=dd.index[:20].tolist()

dailyresults=pd.DataFrame()
for ticker in daily_tickers:
    print(ticker)
    if(data.columns.contains(ticker)):
        y=data[ticker]
    else:

        months=12
        end = datetime.date.today()
        day=end.day
        year=end.year-months//12-1
        month=months%12+1
        start=datetime.datetime(year,month,day)

        y=web.DataReader(ticker,"yahoo",start,end)['Adj Close']
      
    res=forcastStocks(path+'parameters1.csv',ticker,y)
    if(res is not None):
        dailyresults[ticker]=res

dailyresults
```
